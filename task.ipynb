```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the CSV files into DataFrames
train_data = pd.read_csv("Data/train.csv")
```

```python
# Display basic information about the dataset
print(train_data.head())
```

```python
print(train_data.info())
```

```python
print(train_data.describe())
```

```python
# Check for missing values
print(train_data.isnull().sum())
```

```python
# Fill missing values
train_data.dropna(inplace=True)
```

```python
# Check again to confirm missing values are handled
print(train_data.isnull().sum())
```

```python
## check no duplicate values are exist
print(train_data.duplicated().sum())
```

```python
## checking the survival of people 
print(train_data['Survived'].value_counts())
```

```python
#Countplot
sns.countplot(x='Survived',data=train_data,palette='viridis',)
plt.xlabel("Survival status")
plt.ylabel("Number of people")
plt.xticks(ticks=[0,1],labels=['Not survived','survived'])
plt.show()
```

```python
## pie chart
plt.pie(train_data['Survived'].value_counts(), explode=[0,0.04], autopct="%1.2f%%", labels=['Not survived','Survived'])
plt.title("Survival of people")
plt.show()
```

```python
##isualization of people survived from different gender
print(train_data['Sex'].unique())
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.catplot(x='Sex', data=train_data, hue='Sex', palette='viridis', kind='count', legend=False)  # Assigning x variable to hue
plt.xlabel("Gender")
plt.ylabel("Number of people")
plt.show()
```

```python
# countplot survived , not survived
sns.countplot(x='Survived',hue='Sex',data=train_data,palette='viridis')
plt.xlabel("Survival status")
plt.ylabel("Number of people")
plt.xticks(ticks=[0,1],labels=[' Not survived','survived'])
plt.show()
```

```python
## male survival rate
train_data[train_data['Sex'] == 'male'].Survived.groupby(train_data.Survived).count().plot(kind='pie',
figsize=(3, 6),explode=[0,0.05],autopct='%1.1f%%',labels=["Not survived","Survived"])
plt.ylabel("")
plt.title("Male survival rate")
plt.show()
```

```python
## Femal survival rate
train_data[train_data['Sex'] == 'female'].Survived.groupby(train_data.Survived).count().plot(kind='pie',figsize=(3,6),explode=[0,0.05],autopct='%1.1f%%',labels=["Not survived","Survived"])
plt.ylabel("")
plt.title("Female survival rate")
plt.show()
```

```python
# Define cut points and label names
cut_points = [ 0, 5, 12, 18, 35, 60, 100]
label_names = [ 'Infant', "Child", 'Teenager', "Young Adult", 'Adult', 'Senior']
```

```python
# Create the "Age_categories" column
train_data['Age_categories'] = pd.cut(train_data['Age'], bins=cut_points, labels=label_names,right=False)
```

```python
# Creating a pivot table for survival rates based on age categories
age_cat_pivot = train_data.pivot_table(index="Age_categories", values="Survived")
```

```python
# Define colors for each bar
colors = ['blue', 'gray', 'green', 'purple', 'pink', 'brown']
```

```python
# Plotting the bar chart with different colors for each bar
fig, ax = plt.subplots()
bars = ax.bar(age_cat_pivot.index, age_cat_pivot['Survived'], color=colors)
```
